# **Recursive function**

```python
#(function calling itself called recursive function)

def fun(a):
    if a==1:
        return 1
    return a+fun(a-1)

ans= fun(4)

print(ans)
```

```python
def fun(n):
     if n>0:
        fun(n-1)
        print(n,end=" ")
        fun(n-1)

ans=fun(3)
print(ans)
```
