#1.   Arithmetic<br>
    / (True division) ,+ , -, \* , // (floor division) ,% (modulo) ,** power<br>



---
Operator Precedence => Associativity Rule <br>
1) () L-to-R          <br>
2) ** R-TO-L            <br>
3) * , / , % , // L-to-R <br>
4) + , -  L-to-R        <br>








```python
a,b=map(int,input().split())
print(a,"+",b,'=',a+b)
print(f' {a} + {b} = {a+b}')
print(f' {a} - {b} = {a-b}')
print(f' {a} * {b} = {a*b}')
print(f' {a} / {b} = {a/b}')
print(f' {a} // {b} = {a//b}')
print(f' {a} % {b} = {a%b}')
print(f' {a} ** {b} = {a**b}')
```

# 2. Relational Operator <br>
 < , > , <= , >= , ==
  (Equal) ,  != (Not equal)

---
Relational Operator: Used for comparision of data.<br>
At an instance, can write only one condition <br>


---


Operator precedence <br>

1. Arithmetic
2. Relational



```python
a,b,c=map(int,input().split())
print(f' {a} < {b} : {a<b}')
print(f' {b} > {c} : {b>c}')
print(f' {a} <= {c} : {a<=c}')
print(f' {b} >= {a} : {b>=a}')
print(f' {c} == {a} : {c==a}')
print(f' {a} != {b} : {a!=b}')
```

```python
a,b,c=map(int,input().split())# 11 12 13
print(f' {a}+1 < {b} : {a+1<b}')# False
print(f' {b} > {c}-1 : {b>c-1}')#  False
print(f' {a} <= {c}-2 : {a<=c-2}')# True
print(f' {b}-1 >= {a} : {b-1>=a}')# True
print(f' {c}+0 == {a} : {c+0==a}')# False
print(f' {a} != {b}+2 : {a!=b+2}')# True
```

# 03.Logical Operator <br>
 and , or , not <br>

 and => All condition in the given expresion must be true
 <br>

 or => Atleast one condtion must be true in the given expresion


 not => Inverse (True->False) <br>
 (False->True)


---
1. At an instance, can check more than one condition.

or
<br>
Multiple conditions can be written in a single instance.

2. Short circuit Evaluation
---
Operator precedence <br>
1. Arithmetic
2. Relational
3. Logical



```python
a=int(input())
print(f' {a}>=5 and {a}<=20 : {a>=5 and a<=20}')
print(f' {a}>=5 or {a}<=20 : {a>=5 or a<=20}')
print(f' not({a}>=5 and {a}<=20) : {not(a>=5 and a<=20)}')
```

**Q9. Print the following series** <br>
Test case 1: <br>
Input: 7 <br>
Output: 7 5 8 6 9 7 10 <br><br>

Test case 2: <br>
Input: 5 <br>
Output: 5 3 6 4 7




```python
n = int(input())

pos = []
neg = []

for i in range(n):
    pos.append(n + i)

for i in range(n // 2):
    neg.append(n - 2 + i)

res = []
for i in range(n):
    if i % 2 == 0:
        res.append(pos[i // 2])
    else:
        res.append(neg[i // 2])

print(*res)
```
